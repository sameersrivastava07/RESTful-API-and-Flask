```python
# Question:6--->> How do you handle forms in Flask?
# Answer:6-->>>

# Import necessary modules
from flask import Flask, request, render_template_string

# Create Flask app
app = Flask(__name__)

# HTML form template as a string
form_template = """
<!doctype html>
<html>
  <body>
    <h2>Enter your name</h2>
    <form method="POST">
      <input type="text" name="username">
      <input type="submit" value="Submit">
    </form>
    {% if name %}
      <p>Hello, {{ name }}!</p>
    {% endif %}
  </body>
</html>
"""

# Route to display and process the form
@app.route('/', methods=['GET', 'POST'])
def handle_form():
    name = None
    if request.method == 'POST':
        # Accessing the submitted form data
        name = request.form['username']
    return render_template_string(form_template, name=name)

# Run the app
if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# -> request.form is used to get the data from the form when submitted.
# -> Sameer can use this to collect user input, like feedback forms on pwskills website.
# -> render_template_string is used here to keep everything in one file, but normally we use HTML files inside a templates folder.
```

```python
# Question:7--->> How can you validate form data in Flask?
# Answer:7-->>>

# Import necessary modules
from flask import Flask, request, render_template_string

# Create Flask app
app = Flask(__name__)

# HTML form template as a string
form_template = """
<!doctype html>
<html>
  <body>
    <h2>Enter your age</h2>
    <form method="POST">
      <input type="text" name="age">
      <input type="submit" value="Submit">
    </form>
    {% if error %}
      <p style="color:red">{{ error }}</p>
    {% elif age %}
      <p>Your age is {{ age }}</p>
    {% endif %}
  </body>
</html>
"""

# Route to handle form submission and validation
@app.route('/', methods=['GET', 'POST'])
def validate_form():
    error = None
    age = None
    if request.method == 'POST':
        input_age = request.form['age']
        if not input_age.isdigit():  # validation: check if age is numeric
            error = "Please enter a valid number"
        else:
            age = int(input_age)
    return render_template_string(form_template, error=error, age=age)

# Run the app
if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# -> We used isdigit() to check if the input is a valid number.
# -> This kind of validation helps avoid errors in logic or database due to bad input.
# -> For example, Sameer can use this in pwskills to validate if someone entered a valid age before saving it.
```

```python
# Question:8--->> How do you manage sessions in Flask?
# Answer:8-->>>

# Import necessary modules
from flask import Flask, session, redirect, url_for, request

# Create Flask app
app = Flask(__name__)

# Set secret key for session encryption
app.secret_key = 'pwskills_secret_key_123'  # This should be kept secret in real applications

# Route to set session (login simulation)
@app.route('/login', methods=['POST', 'GET'])
def login():
    if request.method == 'POST':
        user = request.form['username']
        session['user'] = user  # Store user in session
        return f"Hello {user}, you are now logged in!"
    return '''
        <form method="post">
            <p><input type=text name=username placeholder='Enter your name'>
            <p><input type=submit value=Login>
        </form>
    '''

# Route to check session (user dashboard simulation)
@app.route('/dashboard')
def dashboard():
    if 'user' in session:
        return f"Welcome back, {session['user']}!"
    return redirect(url_for('login'))

# Route to remove session (logout simulation)
@app.route('/logout')
def logout():
    session.pop('user', None)  # Remove user from session
    return "You have been logged out."

# Run the app
if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# -> Flask’s session stores data across requests using cookies.
# -> A secret key is needed to sign the session cookie securely.
# -> Example: Sameer logs in to pwskills, his username gets stored in the session, and he stays logged in during the visit.
```

```python
# Question:11--->> How do you structure a Flask app using Blueprints?

# Answer:11-->>>

from flask import Flask, Blueprint

# Create a Blueprint for user-related routes
user_bp = Blueprint('user_bp', __name__)

@user_bp.route('/profile')
def profile():
    return "Hello from Sameer's user profile!"

# Create the main Flask app
app = Flask(__name__)

# Register the Blueprint with the app
app.register_blueprint(user_bp, url_prefix='/user')

@app.route('/')
def home():
    return "Welcome to pwskills main page!"

if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# ---> Blueprints help us organize a Flask app into smaller parts or modules.
# ---> This way, we can group related routes and functions separately.
# --->>> For example, we create a 'user_bp' Blueprint for user pages like profile.
# --> Then we register this Blueprint in the main app with a URL prefix like '/user'.
# --->> This makes the app easier to manage and scale, especially in big projects.
```

```python
# Question:12--->> How do you define a custom Jinja filter in Flask?

# Answer:12-->>>

from flask import Flask

app = Flask(__name__)

# Define a custom filter function to reverse a string
def reverse_string(s):
    return s[::-1]

# Register the custom filter with Flask's Jinja environment
app.jinja_env.filters['reverse'] = reverse_string

@app.route('/')
def home():
    # Example usage in template (in real app, use render_template)
    return "Using custom filter: " + reverse_string("pwskills")

if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# ---> Custom Jinja filters let us add new functions to modify data inside HTML templates.
# ---> We write a Python function (like reverse_string) and register it with Flask.
# --->>> Then inside the HTML template, we can use this filter to change data display.
# --> For example, Sameer can use the 'reverse' filter to show strings backward in templates.
# --->> This makes template rendering more powerful and flexible.
```

```python
# Question:13--->> How can you redirect with query parameters in Flask?

# Answer:13-->>>

from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/redirect-example')
def redirect_example():
    # We want to redirect to 'target' route with query parameters like ?name=Sameer&course=pwskills
    return redirect(url_for('target', name='Sameer', course='pwskills'))

@app.route('/target')
def target():
    return "This is the target page after redirect with query parameters."

if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# ---> To redirect with query parameters, we use Flask’s url_for function.
# ---> We pass the route name and query params as keyword arguments.
# --->>> Then we use redirect() to send user to that URL.
# --> For example, Sameer can redirect users with their name and course info as query params.
# --->> This helps pass extra data during redirection in Flask apps.
```

```python
# Question:14--->> How do you return JSON responses in Flask?

# Answer:14-->>>

from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/json-example')
def json_example():
    # We create a Python dictionary to send as JSON response
    data = {
        "name": "Sameer",
        "course": "pwskills",
        "status": "active"
    }
    # ---> We use Flask's jsonify() function to convert dictionary into JSON response
    # --->>> This sets the correct headers automatically for JSON content
    return jsonify(data)

if __name__ == '__main__':
    app.run(debug=True)

# Notes:
# ---> Returning JSON responses is easy with jsonify() in Flask.
# ---> It converts Python dict or list into JSON format for API responses.
# --> We can use this for sending data to clients in RESTful APIs.
# --->> For example, Sameer’s API returns user info as JSON using jsonify().
```

```python
# Question:15--->> How do you capture URL parameters in Flask?

# Answer:15-->>>

from flask import Flask

app = Flask(__name__)

# ---> We define route with variable parts using <variable_name> in URL
@app.route('/user/<username>')
def show_user_profile(username):
    # ---> The 'username' parameter from URL is passed to this function
    # --->>> We can use it to display or process user-specific info
    return f"Hello, {username}! Welcome to pwskills."

# ---> For example, if we visit /user/Sameer, it will capture 'Sameer' as username
# --> Flask automatically captures URL parts as function parameters
# --->> This helps us make dynamic web pages or APIs based on URL inputs

if __name__ == '__main__':
    app.run(debug=True)
```
